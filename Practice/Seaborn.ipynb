```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.get_dataset_names()
```

```python
tips = sns.load_dataset("tips")
iris = sns.load_dataset("iris")
titanic = sns.load_dataset("titanic")
planets = sns.load_dataset("planets")
```

# Tips

```python
tips.head()
```

```python
sns.scatterplot(data=tips,
                x="tip",
                y="total_bill",
                hue="day",
                size="size",
                palette="YlGnBu")
plt.show()
```

```python
plt.figure(figsize=(4,3))
sns.histplot(data=tips,
             x="tip",
             kde=True,)
plt.show()
```

```python
sns.barplot(data=tips,
            x="sex",
            y="tip",
            palette="YlGnBu")
```

```python
sns.boxenplot(data=tips,
              x="day",
              y="tip",
              hue="sex",
              palette="YlGnBu")
```

```python
sns.boxenplot(data=tips,
              x="day",
              y="total_bill",
              hue="sex",
              palette="YlGnBu")
```

```python
sns.stripplot(data=tips,
              x="day",
              y="tip",
              hue="sex",
              palette="YlGnBu",
              dodge=True
              )
```

```python
sns.jointplot(data=tips,
              x="tip",
              y="total_bill",
              hue="sex")
plt.show()
```

```python
sns.jointplot(data=tips,
              x="tip",
              y="total_bill",
              kind="kde",
              fill=True,
              cmap="YlGnBu")
plt.show()
```

```python
sns.pairplot(data=titanic.select_dtypes(["number"]),
             hue="pclass")
plt.show()
```
