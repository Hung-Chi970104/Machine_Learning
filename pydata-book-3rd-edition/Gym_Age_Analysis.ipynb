```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
gym_dataset = pd.read_csv("pydata-book-3rd-edition/archive/gym_members_exercise_tracking.csv")
gym_dataset
```

```python
gym_dataset = gym_dataset[gym_dataset.Workout_Type == "Strength"]
gym_dataset
```

```python
sorted_data = gym_dataset.sort_values(by=["Age", "Gender"], ascending=[True, False])
sorted_data
```

```python
summary_df = sorted_data.groupby(["Age", "Gender"])[["Max_BPM", "Avg_BPM", "Resting_BPM"]].mean().reset_index()
summary_df = summary_df.round(1)
summary_df
```

```python
male_bpm = summary_df[summary_df.Gender == "Male"]
male_bpm.head()
```

```python
male_bpm.plot(x="Age", y=["Max_BPM", "Avg_BPM", "Resting_BPM"])
```

```python
summary_df.to_csv("Age-BPM.csv")
```
